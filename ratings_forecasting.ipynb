## Data transformation for ML models

```python
YEAR_QUARTERS = [1,1,1,2,2,2,3,3,3,4,4,4]
# df_for_ml_model = pd.read_csv('data/banijay_op1data_content_ratings_twitter_for_model.csv',compression='zip')
def get_the_quarter(x):
    num = 0
    if x[0] == '0':
        num = int(x[-1])
    else:
        num = int(x)
    return YEAR_QUARTERS[num-1]


df_for_ml_model['year_quarter'] = df_for_ml_model['date'].apply(lambda x: get_the_quarter(str(x)[3:5]))

df_for_ml_model
```

## Training the Linear Rergession Model

```python
Y = df_for_ml_model['Kdh000']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25)

model = LinearRegression()
model.fit(X_train,y_train)


import seaborn as sns
import sklearn.metrics as sm

predictions = model.predict(X_test)


predictions_on_seen_data = model.predict(X_train)
```

## Assessing the accuracy of the model

### Correlation of the predicted rating and actual rating based on the test and train data

```python
np.corrcoef(y_test,predictions)[0,1]

np.corrcoef(y_train,predictions_on_seen_data)
```

### Comparison of the model metrics based on the test and train data

```python
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, predictions), 2))
print("Mean squared error =", round(sm.mean_squared_error(y_test, predictions), 2))
print("Median absolute error =", round(sm.median_absolute_error(y_test, predictions), 2))
print("Explain variance score =", round(sm.explained_variance_score(y_test, predictions), 2))
print("R2 score =", round(sm.r2_score(y_test, predictions), 2))

print("\n")

print("Mean absolute error =", round(sm.mean_absolute_error(y_train, predictions_on_seen_data), 2))
print("Mean squared error =", round(sm.mean_squared_error(y_train, predictions_on_seen_data), 2))
print("Median absolute error =", round(sm.median_absolute_error(y_train, predictions_on_seen_data), 2))
print("Explain variance score =", round(sm.explained_variance_score(y_train, predictions_on_seen_data), 2))
print("R2 score =", round(sm.r2_score(y_train, predictions_on_seen_data), 2))
```

### Estimating the coefficients of a linear regression model using the normal equations.

```python
def LinReg(X, y):

    XtX_inverse = X.T.dot(X)
    XtY = X.T.dot(y)

    return np.linalg.solve(XtX_inverse,XtY)


LinReg(X,Y)
```

## Training the Decision Tree Model

```python
from sklearn.tree import DecisionTreeRegressor

X_train_tree, X_test_tree, y_train_tree, y_test_tree = train_test_split(X, Y, test_size=0.25, random_state= 1)

clf = DecisionTreeRegressor(max_depth=2)

clf = clf.fit(X_train_tree,y_train_tree)

y_pred_tree = clf.predict(X_test_tree)
```

### Calculating the model metrics based on the test data

```python
print("Mean absolute error =", round(sm.mean_absolute_error(y_test_tree, y_pred_tree), 2))
print("Mean squared error =", round(sm.mean_squared_error(y_test_tree, y_pred_tree), 2))
print("Median absolute error =", round(sm.median_absolute_error(y_test_tree, y_pred_tree), 2))
print("Explain variance score =", round(sm.explained_variance_score(y_test_tree, y_pred_tree), 2))
print("R2 score =", round(sm.r2_score(y_test_tree, y_pred_tree), 2))

print("Correlation coefficient of predicted ratings and actual ratings =", np.corrcoef(y_test_tree,y_pred_tree)[0,1])
```

### Visualising the Decision Tree

```python
from sklearn import tree

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf,feature_names=X.columns.tolist(),class_names=Y,filled=True)
plt.savefig("decision_tree.png", dpi = 600)
```

### Tuning the hyperparameter of the Decision Tree by changing the max depth of it

```python
from sklearn.model_selection import GridSearchCV

parameters = {"splitter":["best","random"],"max_depth" : [2,4,8,16,32],"min_samples_leaf":[2,4]}

grid_search = GridSearchCV(clf, parameters, cv = 5)


grid_search.fit(X_train_tree,y_train_tree)

print("Best parameters: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
```

```python
tree_model_tunned = DecisionTreeRegressor(splitter="best",max_depth=32,min_samples_leaf=2)
tree_model_tunned.fit(X_train_tree,y_train_tree)

tree_tuned_predicts = tree_model_tunned.predict(X_test_tree)

tree_model_tunned.score(X_test_tree,y_test_tree)
```
